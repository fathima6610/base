```python
from transformers import pipeline, AutoModelForSequenceClassification,AutoTokenizer
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
pipe = pipeline('sentiment-analysis')
```

```python
import csv
import random
from transformers import pipeline

# Create a sentiment analysis pipeline
sentiment_analysis_pipeline = pipeline("sentiment-analysis")

# Path to your CSV file
csv_file_path = "/content/rnn_analysis.csv"

# Function to read CSV and perform sentiment analysis on random text
def perform_sentiment_analysis_on_random_text(csv_file_path, text, num_samples):
    with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        lines = [row[column_name] for row in reader]
        selected_texts = random.sample(lines, min(num_samples, len(lines)))
        for text in selected_texts:
            result = sentiment_analysis_pipeline(text)
            sentiment_label = result[0]["label"]
            sentiment_score = round(result[0]["score"], 2)
            print(f"Text: {text}")
            print(f"Sentiment: {sentiment_label}")
            print(f"Confidence Score: {sentiment_score}")
            print("-----------------------------")

# Name of the column containing text
column_name = "text"

# Number of random samples you want to select
num_samples = 3

# Call function to perform sentiment analysis on random text from CSV
perform_sentiment_analysis_on_random_text(csv_file_path, column_name,num_samples)
```
